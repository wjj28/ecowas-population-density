### ECOWAS
The Economic Community of West African States (ECOWAS) is made up of fifteen member countries that are located in the Western African region.

```python
ECOWAS = ['Benin','Burkina Faso','Cabo Verde', 'Cote d\'Ivoire','Gambia, The',
         'Ghana', 'Guinea','Guinea-Bissau','Liberia','Mali','Niger','Nigeria','Senegal','Sierra Leone','Togo']
```

To make sure that every single state member of ECOWAS is in the list, and to print the ones we found

```python
countries_name = countries['Country Name']

i = set.intersection(set(countries_name),set(ECOWAS))
if len(i)>0:
    print("duplicates found",i)
```

now we want to select only the ECOWAS state members

```python
western_africa = (countries[countries_name.isin(ECOWAS)])
```

```python
western_africa
```

```python
top10 = western_africa.sort_values('2020', ascending = True).head(15)
plt.xticks(rotation = '45')
plt.title('Population Density of ECOWAS members for the year 2020', color = 'r')
plt.rcParams['figure.figsize'] = (15, 15)
plt.rcParams['font.size'] = 15
plt.xlabel('Countries', color = 'r')
plt.ylabel('Population Density', color = 'g')
plt.bar(top10['Country Code'],
        top10['2020'],
        color = cm.rainbow(np.linspace(0, 1, len(top10))))
```

```python
countries = western_africa['Country Name']
populationDensity2020 = western_africa['2020']
plt.rcParams['figure.figsize'] = (20, 20)
plt.rcParams['font.size'] = 20
plt.xticks(rotation = '90')
plt.bar(countries, populationDensity2020, color = cm.rainbow(np.linspace(0, 1, len(countries))))
```

# Range of Population Density for the years 1962-2020
 Population density data can be used to quantify demographic information and to assess relationships with ecosystems, human health, and infrastructure.

```python
minimum = western_africa.loc[:, western_africa.columns != 'Country Name'].min()
maximum = western_africa.loc[:, western_africa.columns != 'Country Name'].max()
diff = maximum[1:] - minimum[1:]
minOfMax = maximum[1:].min()
plt.title('Range of Population Density for the years 1962-2020', color ='g')
plt.rcParams['figure.figsize'] = (15, 20)
plt.rcParams['font.size'] = 20
plt.xticks(rotation = '90')
plt.xlabel('Years', color='g')
plt.ylabel('Population Density', color = 'g')
plt.bar(western_africa.columns[2:], diff.apply(lambda x: x-minOfMax), color = cm.rainbow(np.linspace(0, 1, western_africa.shape[1])))
```
